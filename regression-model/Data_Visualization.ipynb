```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../neural-network/The FINAL Dataset.csv')
display(df)
```

This log scaling of the # of cases was added after noticing outliers in the data. 

```python
df['cases_log'] = np.log10(df['cases'] + 1)
df.hist(column='cases_log')
```

```python
corr = df.corr()
corr.style.format(precision = 4)
#sns.heatmap(tips_df.corr(), annot = True)
sns.heatmap(corr, annot = True, cmap = 'coolwarm', square = True)
```

```python
sns.pairplot(df)
plt.scatter(df['cases_log'], df['percent change 2020'], s = 1)
```

```python
df['cases_rate'] = df['cases']/df['2020 population']
df.hist(column='cases_rate')
```

```python
plt.xlim(0, 0.15)
plt.ylim(-20, 20)
plt.scatter(df['cases_rate'], df['percent change 2020'], s = 1)
```
