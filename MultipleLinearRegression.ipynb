```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk

# Use this instead
df = pd.read_csv('The FINAL Dataset.csv')


from sklearn import linear_model

df =  df.dropna()

df['2019 raw GDP'] = df['2019 raw GDP'].astype('float64')
df['percent change 2020'] = df['percent change 2020'].astype('float64')
df['cases'] = df['cases'].astype('int32')
df['deaths'] = df['deaths'].astype('int32')

from sklearn.preprocessing import StandardScaler
  
#object= StandardScaler()
  
#df['cases'] = object.fit_transform(df['cases'].values.reshape(-1, 1)) 
#df['deaths'] = object.fit_transform(df['deaths'].values.reshape(-1, 1)) 
#df['2020 population'] = object.fit_transform(df['2020 population'].values.reshape(-1, 1))
#df['2019 raw GDP'] = object.fit_transform(df['2019 raw GDP'].values.reshape(-1, 1)) 
#df['percent change 2020'] = object.fit_transform(df['percent change 2020'].values.reshape(-1, 1)) 

X = df.copy().drop(columns = ['percent change 2020', 'state', 'county'])
y = df['percent change 2020']
display(df)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.2, random_state= 25)


regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)

print(regr.coef_)


predictions = regr.predict(x_test)

from sklearn.metrics import classification_report
#print(classification_report(y_test, predictions))

from sklearn.metrics import mean_squared_error
print("Mean Squared Error: %.7f" % mean_squared_error(y_test, predictions))

from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```
